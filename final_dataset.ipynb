# Merged Dataset

### Load the Dataset

```python
merged_df = pd.read_csv(r'merged_dataset/cleaned_merged_dataset.csv')
```

### Display first few rows

```python
merged_df.head()
```

### Check Datatypes of Dataset

```python
merged_df.dtypes
```

### Convert 'dteday' column to datetime

```python
merged_df['dteday'] = pd.to_datetime(merged_df['dteday'], format='%Y-%m-%d')
```

### Merge datasets on 'instant'

* Both datasets have the same column names, causing pandas to append _x and _y suffixes to distinguish them after merging. This happens when both datasets contain the same columns but may have different values.

```python
final_merge = pd.merge(merged_df, df3, on='instant', how='outer')
```

### Removes _y columns (keeping the first datasets values).

```python
columns_to_drop = [col for col in final_merge.columns if col.endswith('_y')]
final_merge.drop(columns=columns_to_drop, inplace=True)
```

### Renames _x columns back to their original names.

```python
final_merge.rename(columns=lambda x: x.rstrip('_x'), inplace=True)
```

### Dimention of Final Merge Dataset

```python
final_merge.shape
```

### Display first few rows

```python
final_merge.head()
```

### Check Datatypes of Dataset

```python
final_merge.dtypes
```

### Identify Unique Values

```python
final_merge.nunique()
```

### Check for Missing Values

```python
final_merge.isnull().sum()
```

### Fill Missing Values

#### Fill missing values for numerical columns with mean

```python
final_merge.fillna(final_merge.select_dtypes(include=['number']).median(), inplace=True)
```

#### Fill missing values for categorical columns with mode

```python
# Fill missing values with mode for object columns in a safe way
final_merge.fillna({col: final_merge[col].mode()[0] for col in final_merge.select_dtypes(include=['object']).columns}, inplace=True)

final_merge = final_merge.infer_objects(copy=False)
```

### Check if there are any null values in the dataset summary

```python
print(final_merge.info())
```

```python
final_merge.describe()
```

### Check for Outliers

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=final_merge.select_dtypes(include=['number']))
plt.title('Boxplot for Outlier Detection')
plt.show()
```

### Handling Outliers using IQR Method

```python
# Handling Outliers using IQR Method (Capping instead of Removing)

numeric_cols = final_merge.select_dtypes(include=['number']).columns

Q1 = final_merge[numeric_cols].quantile(0.25)
Q3 = final_merge[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Capping outliers instead of removing them (Using .loc to avoid SettingWithCopyWarning)
final_merge.loc[:, numeric_cols] = final_merge[numeric_cols].clip(lower=lower_bound, upper=upper_bound, axis=1)
```

### Re-check for outliers after handling them

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=final_merge.select_dtypes(include=['number']))
plt.title('Boxplot After Outlier Handling')
plt.show()
```
