# Data Visualization on Dataset 1

### 1. Temprature Destribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df1['temp'], bins=30, kde=True)
plt.title("Temperature Distribution in Dataset 1")
plt.xlabel("Temperature (Normalized)")
plt.ylabel("Frequency")
plt.show()
```

### Insights

* This plot represents the distribution of temperature in the dataset.
* KDE (Kernel Density Estimation) curve is overlaid to visualize the density of the data points.
* Most temperatures are concentrated in a specific range (between 0.15 to 0.25 normalized values).
* Few extreme temperatures (either very low or very high) are observed.

### 2. Hourly Distribution of Bike Rentals

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='hr', data=df1, hue='hr', palette='viridis', legend=False)
plt.title('Hourly Distribution of Bike Rentals')
plt.xlabel('Hour of the Day')
plt.ylabel('Count')
plt.show()
```

### Insights

* Morning peak (8 AM) and evening peak (5-6 PM) reflect commuter behavior.
* Late-night hours (0-5 AM) have minimal activity.

### 3. Rentals by Day of the Week

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='weekday', data=df1, hue='weekday', palette='magma', legend=False)
plt.title('Rentals by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Count')
plt.show()
```

### Insights

* Weekdays (Monday-Friday) show higher rentals, reflecting work commutes.
* Weekends have fewer rentals, indicating leisure-based behavior.

### 4. Temperature vs. Hourly Rentals

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='temp', y='hr', data=df1, alpha=0.7, color='purple')
plt.title('Temperature vs. Hourly Bike Rentals (Dataset 1)')
plt.xlabel('Temperature (Normalized)')
plt.ylabel('Hour of the Day')
plt.grid(True)
plt.show()
```

### Insights

* Higher temperatures align with increased rentals during the daytime.
* Cooler temperatures correlate with lower rentals.

### Export cleaned dataset to CSV

```python
df1.to_csv(r"cleaned_dataset/cleaned_dataset_1.csv", index=False)
print("Cleaned dataset exported successfully as 'cleaned_dataset1.csv'")
```

# Dataset 2

### load the dataset

```python
df2 = pd.read_csv(r"original_dataset/dataset_2.csv")
```

### Display the first few rows of the dataset

```python
display(df2.head())
```

### Check the dimensions of the dataset

```python
print(f"Dataset Shape: {df2.shape}")
```

### Check data types of the dataset

```python
display(df2.dtypes)
```

### Drop unnecessary columns 'Unnamed: 0'

```python
unnecessary_columns = ['Unnamed: 0']
df2 = df2.drop(columns=[col for col in unnecessary_columns if col in df2.columns], errors='ignore')
```

### Printing dataset after drop unnecessary column

```python
display(df2.head())
```

### Summary of datatype counts

```python
print("\nSummary of data types:\n",df2.dtypes.value_counts())
```

### Checking for missing values

```python
print("\nMissing Values Count:",df2.isnull().sum())
```

### Handle missing values

```python
df_numeric = df2.select_dtypes(include=['float64', 'int64'])
df2[df_numeric.columns] = df_numeric.fillna(df_numeric.median())
```

### Validate correctness of data

```python
print("\nData Sample After Cleaning:\n")
display(df2.head())
```

### Summary statistics for all numerical columns

```python
df2.describe()
```

# Data Visualization on Dataset 2

### 1. Bike Rentals vs Temperature

```python
if 'atemp' in df2.columns and 'cnt' in df2.columns:
    plt.figure(figsize=(8,5))
    sns.scatterplot(x=df2['atemp'], y=df2['cnt'], alpha=0.5)
    plt.xlabel('Temperature')
    plt.ylabel('Bike Rentals')
    plt.title('Bike Rentals vs Temperature')
    plt.show()
```

### Insights

* Displays the relationship between temperature and bike rentals.
* Each point represents a data entry of temperature and corresponding rentals.
*  As temperature increases, bike rentals tend to increase.
*  Some points suggest low rentals even in moderate temperatures (could indicate bad weather or holidays).

### 2. Pair Plot - Pairplot to analyze relationships between multiple variables

```python
sns.pairplot(df2[['atemp', 'hum', 'windspeed', 'cnt']])
plt.show()
```

### Insights

* Pairwise relationship across multiple numerical columns (e.g., temperature, humidity, windspeed, and total rentals).
* Temperature and atemp (feeling temperature) show a strong correlation with bike rentals.
* Humidity and windspeed exhibit a weaker correlation, implying less influence on demand.

### 3. Relationship Between Humidity and Rentals

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='hum', y='cnt', data=df2, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Humidity vs. Bike Rentals (Dataset 2)')
plt.xlabel('Humidity (Normalized)')
plt.ylabel('Total Rentals')
plt.grid(True)
plt.show()
```

### Insights

* Rentals decrease as humidity increases.
* Users prefer renting bikes during low humidity days.

### 4. Wind Speed Impact on Casual vs. Registered Users

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='windspeed', y='casual', data=df2, color='blue', label='Casual Users')
sns.scatterplot(x='windspeed', y='registered', data=df2, color='red', label='Registered Users')
plt.title('Impact of Wind Speed on Users (Dataset 2)')
plt.xlabel('Wind Speed (Normalized)')
plt.ylabel('User Count')
plt.legend()
plt.grid(True)
plt.show()
```

### Insights

* asual users drop as wind speed increases.
* Registered users are less affected by wind speed.
* Low wind speeds see more rentals from both groups.
* Casual users are more sensitive to wind changes.

### 5. Registered vs. Casual User Comparison

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(df2['casual'], label='Casual Users', fill=True, color='blue')
sns.kdeplot(df2['registered'], label='Registered Users', fill=True, color='red')
plt.title('Distribution of Casual vs Registered Users (Dataset 2)')
plt.xlabel('User Count')
plt.legend()
plt.show()
```

### Insights

* Registered users form the majority of rentals.
* Casual users exhibit more variation and sporadic behavior.

### Export cleaned dataset to CSV

```python
df2.to_csv("cleaned_dataset/cleaned_dataset_2.csv", index=False)
print("Cleaned dataset exported successfully as 'cleaned_dataset_2.csv'")
```

# Data Visualization on Merged Dataset

### 1. Bike Rentals Over Time

```python
plt.figure(figsize=(12, 6))
plt.plot(merged_df['dteday'], merged_df['cnt'], marker='o', linestyle='-', color='green')
plt.title('Bike Rentals Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Total Bike Rentals')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### Insights

* Temporal trends of total bike rentals over the given timeframe.
* Shows spikes and dips in rental counts across days.
* Repeated patterns suggest cyclical behaviorhigher rentals on specific days.
* Sudden spikes could indicate special events; dips may show holidays or bad weather.

### 2. Pie Chart: Distribution of Weather Conditions

```python
plt.figure(figsize=(6,6))
weather_labels = {1: 'Clear', 2: 'Mist', 3: 'Light Snow', 4: 'Heavy Rain'}
merged_df['weathersit'] = merged_df['weathersit'].map(weather_labels)
merged_df['weathersit'].value_counts().plot.pie(autopct='%1.1f%%', cmap='coolwarm', legend=True)
plt.title('Distribution of Weather Situations (Merged Dataset)')
plt.ylabel('')
plt.legend(title="Weather Conditions")
plt.show()
```

### Insights

* The percentage of bike rentals under different weather situations (e.g., Clear, Mist, Rain).
* Clear Weather accounts for the majority of bike rentals.
* Rain/Snow significantly reduces rentals.

### 3. Hourly Bike Rental Distribution

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='hr', data=merged_df, hue='hr', palette='viridis', legend=False)
plt.title('Hourly Bike Rental Distribution')
plt.xlabel('Hour of the Day')
plt.ylabel('Rental Count')
plt.show()
```

### 4. Bike Rentals by Season

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='season', y='cnt', data=merged_df, hue='season', palette='magma', legend=False)
plt.title('Bike Rentals by Season')
plt.xlabel('Season (1 = Winter, 2 = Spring, 3 = Summer, 4 = Fall)')
plt.ylabel('Bike Rentals')
plt.show()
```

### Insights

* Summer and Fall have higher median rentals.
* Winter shows lower bike rentals.

### 5. Bike Rentals by Season and Hour

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='hr', y='cnt', hue='season', data=merged_df, palette='Set1')
plt.title('Bike Rentals by Hour and Season')
plt.xlabel('Hour of the Day')
plt.ylabel('Bike Rentals')
plt.legend(title='Season (1 = Winter, 2 = Spring, 3 = Summer, 4 = Fall)')
plt.show()
```

### Insights

* Morning (8 AM) and Evening (5 PM) peaks are consistent across seasons.
* Summer shows higher rentals across most hours.

### 6. Stacked Area Plot  Casual vs. Registered Users Over Time

```python
plt.figure(figsize=(12, 6))
plt.stackplot(merged_df['dteday'], merged_df['casual'], merged_df['registered'], labels=['Casual Users', 'Registered Users'], colors=['#1f77b4', '#ff7f0e'])
plt.title('Casual vs Registered Users Over Time')
plt.xlabel('Date')
plt.ylabel('Bike Rentals')
plt.legend()
plt.grid(True)
plt.show()
```

### Insights

* Registered users dominate overall bike rentals.
* Casual users show seasonal spikes, especially during holidays.

### 7. Hexbin Plot  Temperature vs. Bike Rentals

```python
plt.figure(figsize=(10, 6))
plt.hexbin(merged_df['temp'], merged_df['cnt'], gridsize=30, cmap='plasma')
plt.colorbar(label='Density of Rentals')
plt.title('Temperature vs. Bike Rentals (Density Plot)')
plt.xlabel('Temperature (Normalized)')
plt.ylabel('Bike Rentals')
plt.show()
```

### Insights

* Moderate temperatures (around 0.2-0.3) drive the most rentals.
* Extreme temperatures (too hot or cold) reduce bike usage.

### 8. Violin Plot  User Behavior by Hour

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='hr', y='cnt', data=merged_df, hue='holiday', split=True, palette='viridis')
plt.title('Hourly Bike Rentals (Holiday vs Non-Holiday)')
plt.xlabel('Hour of the Day')
plt.ylabel('Total Rentals')
plt.legend(title='Holiday (1=Yes, 0=No)')
plt.show()
```

### Insights

* Non-holidays show peaks during 8 AM and 5 PM (commute hours).
* Holidays have smoother patterns without sharp spikes.

### 9. Radial Bar Plot  Hourly Bike Rentals

```python
angles = np.linspace(0, 2 * np.pi, 24, endpoint=False).tolist()
rental_by_hour = merged_df.groupby('hr')['cnt'].mean().values

# Create the plot
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
bars = ax.bar(angles, rental_by_hour, color='coral', alpha=0.8, label='Average Bike Rentals')

# Add legend
ax.legend(loc='upper right', bbox_to_anchor=(1.2, 1.1))

# Customize the plot
ax.set_xticks(angles)
ax.set_xticklabels(range(24))
ax.set_title('Radial Bar Plot: Hourly Bike Rentals', va='bottom')

plt.show()
```

### Insights

* 8 AM and 5 PM are the most active hours (commute peaks).
* Early mornings (0-5 AM) have minimal rentals.

### Export cleaned merged dataset to CSV

```python
merged_df.to_csv(r"merged_dataset/cleaned_merged_dataset.csv", index=False)
print("Cleaned merged dataset exported successfully as 'cleaned_merged_dataset.csv'")
```
